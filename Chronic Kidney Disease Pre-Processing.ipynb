```python
from sklearn.preprocessing import LabelEncoder
lb=LabelEncoder()
dataset['rbc']=lb.fit_transform(dataset['rbc'])
dataset['pc']=lb.fit_transform(dataset['pc'])
dataset['pcc']=lb.fit_transform(dataset['pcc'])
dataset['ba']=lb.fit_transform(dataset['ba'])
dataset['htn']=lb.fit_transform(dataset['htn'])
dataset['dm']=lb.fit_transform(dataset['dm'])
dataset['cad']=lb.fit_transform(dataset['cad'])
dataset['appet']=lb.fit_transform(dataset['appet'])
dataset['pe']=lb.fit_transform(dataset['pe'])
dataset['ane']=lb.fit_transform(dataset['ane'])
dataset['classification']=lb.fit_transform(dataset['classification'])
```

```python
print(dataset)
```

```python
dataset['age']=dataset['age'].fillna(dataset['age'].mean())
```

```python
dataset.replace('\t?', float('nan'), inplace=True)
```

```python
columns_to_convert = [ 'bp','sg',   'al' ,  'su',  'rbc',  'pc',  'pcc' , 'ba'  ,'bgr', 'bu', 'sc', 'sod', 'pot' ,'hemo' ,'pcv' , 'wc' , 'rc' ,'htn',  'dm'  ,'cad',  'appet' , 'pe' , 'ane' , 'classification']
```

```python
for column in columns_to_convert:
    dataset[column] = pd.to_numeric(dataset[column], errors='coerce')
```

```python
dataset.dropna(inplace=True)
```

```python
dataset=pd.DataFrame(dataset,columns=['id',   'age',   'bp',     'sg',   'al' ,  'su',  'rbc',  'pc',  'pcc' , 'ba'  ,'bgr', 'bu', 'sc', 'sod', 'pot' ,'hemo' ,'pcv' , 'wc' , 'rc' ,'htn',  'dm'  ,'cad',  'appet' , 'pe' , 'ane' , 'classification'])
```
